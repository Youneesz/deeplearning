```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=100, n_features=2, centers=2, random_state=0)
y = y.reshape(y.shape[0], 1)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
W1 = np.linspace(-10, 10, 100)
```

```python
W1
```

```python
W2 = np.linspace(-10, 10, 100)
W2
```

```python
W11, W22 = np.meshgrid(W1, W2)
```

```python
W11.shape
```

```python
W22.shape
```

```python
b = 0
W_final = np.c_[W11.ravel(), W22.ravel()].T
W_final.shape
Z = X.dot(W_final) + b 
```

```python
A = 1 / (1 + np.exp(-Z))
```

```python
A.shape
```

```python
epsilon = 1e-15
L = -1/y.shape[0] * np.sum(y*np.log(A + epsilon) + (1-y)*np.log(1-A + epsilon), axis=0).reshape(W11.shape)
L.shape
```

```python
plt.contourf(W11, W22, L, cmap='magma')
plt.colorbar()
```
